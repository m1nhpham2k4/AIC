```python
import os
import numpy as np
folder = "./features"
all_embeddings = []

for file in os.listdir(folder):
    if file.endswith(".npy"):
        path = os.path.join(folder, file)
        arr = np.load(path)
        all_embeddings.append(arr)
        print(f"Đã load {file}, shape = {arr.shape}")

# --- 3. Ghép lại thành một ma trận ---
embeddings = np.vstack(all_embeddings)
print("Tổng shape embeddings:", embeddings.shape)
```
