```python
# !pip install transformers
# !pip install nltk
# !pip install sentencepiece
# !pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cpu
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize

# Load model
model_name = "Helsinki-NLP/opus-mt-vi-en"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
with open('D:\\AIC\\AIC\\Data_extraction\\text_vi\\L001_V001.txt', 'r', encoding='utf-8') as f:
    text_vi = f.read()
```

```python
text_vi
```

```python
# Nếu bạn muốn chia tốt hơn cho tiếng Việt, dùng đoạn sau thay cho `sent_tokenize`:
import re
sentences = re.split(r', ', text_vi)
```

```python
sentences = [sentence for sentence in sentences if sentence]
sentences
```

```python
# Step 2: Dịch từng câu
translations = []
for sentence in sentences:
    inputs = tokenizer(sentence, return_tensors="pt", truncation=True, max_length=512)
    outputs = model.generate(inputs["input_ids"], num_beams=5, max_length=512, early_stopping=True)
    translated = tokenizer.decode(outputs[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)
    translations.append(translated)

```

```python
translations[:10]
```

```python
# Step 3: In kết quả
translated_text = ', '.join(translations)
print(translated_text)
```

```python
with open("D:\\AIC\\AIC\\Data_extraction\\text_eng\\L01_V001.txt", "w", encoding = 'utf-8') as f:
    f.write(translated_text)
```
